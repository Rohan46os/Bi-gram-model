```python
import requests
import torch
import numpy as np
```

```python
Url = 'https://raw.githubusercontent.com/dominictarr/random-name/refs/heads/master/first-names.txt'
data = requests.get(Url)
```

```python
data = data.text
```

```python
data=data.splitlines()
```

```python
data[:10]
```

```python
new_list = []

for words in data:
  new_words = words+'.'
  new_list.append(new_words)
```

```python
new_list[:5]
```

```python
context_length = 3
i = 0
contexted_words = []
labels = []

for ele in new_list:
  for i in range(len(ele)-3): #for i in range(len(ele) - context_length):
    new_words = ele[i]+ele[i+1]+ele[i+2] # context = ele[i : i + context_length]
    y_val = ele[i+3] #
    labels.append(y_val)
    contexted_words.append(new_words)
```

```python
contexted_words[:5]
```

```python
labels[:5]
```

```python
all_letters = sorted(set(list(''.join(new_list))))
```

```python
all_letters[:5]
```

```python
actual_letters = {}

for index, letters in enumerate(all_letters):
  set_of_letters = {letters: index}
  actual_letters.update(set_of_letters)
```

```python
lookup_table = {s:i for i, s in enumerate(all_letters)}

```

```python
indiv_words = []

for letters in contexted_words:
  hlo = [letters[0], letters[1], letters[2]]
  indiv_words.append(hlo)
```

```python
list(indiv_words[:5])
```

```python
encoded_contexts = []
encoded_labels = []

for word in contexted_words:
    encoded_context = [actual_letters[char] for char in word]
    encoded_contexts.append(encoded_context)

for label in labels:
    encoded_labels.append(actual_letters[label])

display(encoded_contexts[:5])
display(encoded_labels[:5])
```
